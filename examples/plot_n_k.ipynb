## Materials plotting example

JaxLayerLumos can accommodate a wide variety of materials.

```python
from jaxlayerlumos import utils_materials
import matplotlib.pyplot as plt
from jaxlayerlumos import utils_spectra
import jax.numpy as jnp
import numpy as onp

import scipy.constants as scic

from jaxlayerlumos.utils_units import convert_m_to_nm
```

We list all the materials and the file location.  Additional materials and files can be added by the user by changing the file materials.json() and placing the materials n and k parameters in the materials folder.  The wavelengths are given in microns.

```python
materials_dict,_ = utils_materials.load_json()
for key, value in materials_dict.items():
    print(f"{key}: {value}")
```

This plots the data and the extrapolated data.

```python
material = onp.array(['GaInP'])

data_n, data_k = utils_materials.load_material(material[0])

plt.figure(figsize=(8, 6))
utils_spectra.convert_frequencies_to_wavelengths(data_n[:,0])
plt.scatter(convert_m_to_nm(utils_spectra.convert_frequencies_to_wavelengths(data_n[:,0])), data_n[:,1], marker='o')

# plot extrapolated values
wavelengths = jnp.linspace(280e-9, 1000e-9, 1000)
frequencies = scic.c / wavelengths

n_k = utils_materials.get_n_k(material, frequencies)
# n_k.shape
plt.plot(convert_m_to_nm(wavelengths), n_k.real, marker = 'x', color = 'r')
```
